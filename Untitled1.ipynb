```python
import random
import pandas as pd

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})

def one_hot(whoAmI: str, key: str) -> int:
    if whoAmI == key:
      return 1
    else:
      return 0


for i in set(data['whoAmI']):
  data[i] = [ one_hot(j, i) for j in data['whoAmI']]

data.head()
```
